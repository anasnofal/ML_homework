```python
import math

import sympy as sp
```

```python
def function(x_1, x_2,x_3): 

    return math.exp(x_1-3*x_2+3) + math.exp(3*x_2-2*x_3-2)+ math.exp(2*x_3-x_1+2)


def argmin_x1(argmin):
    x1, x2, x3 = argmin
    return (3 * x2 + 2 * x3 - 1) / 2


def argmin_x2(argmin):
    x1, x2, x3 = argmin
    return (x1 + 2 * x3 + 5) / 6


def argmin_x3(argmin):
    x1, x2, x3 = argmin
    return (x1 + 3 * x2 - 4) / 4


def coordinate_descent(f: callable, argmin: list[function], x_t0: list[float], max_iter: int = 25):
    current_value = x_t0.copy()
    for i in range(max_iter):
        for j in range(len(x_t0)):
            current_value[j] =  argmin[j](current_value)
    return current_value
```

```python
coordinate_descent(function, [argmin_x1, argmin_x2, argmin_x3], [1, 20, 5], 19)
```
